```python
import numpy as np
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
from urllib.parse import quote
from selenium import webdriver
```

```python
# Selenium Start

driver = webdriver.Chrome('./chromedriver')
```

```python
# naver crawlering
key_word = input('검색어를입력하세요 : ')

# url 설정
url = f'https://search.naver.com/search.naver?query={key_word}&where=blog&sm=tab_opt'

# requests로 데이터 요청하기
resp = requests.get(url)

# 설정한 url 로 데이터 get 함수 요청
driver.get(url)
```

```python
# 화면상에서 스크롤 끝까지 내리기
SCROLL_PAUSE_SEC = 1

# 스크롤 높이 가져옴
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 끝까지 스크롤 다운
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 1초 대기
    time.sleep(SCROLL_PAUSE_SEC)

    # 스크롤 다운 후 스크롤 높이 다시 가져옴
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
# 접근한 페이지의 소스코드를 텍스트로 저장 (Str)
page_text = driver.page_source

# page_text를 html list로 변환 
html = BeautifulSoup(page_text, 'html.parser')
```

```python
# 제목 title

title = []

for i in html.select('a.api_txt_lines.total_tit') :
    title.append(i.text)
    
print(title)
```

```python
# blog_url
blog_url=[]

for i in html.select('a.api_txt_lines.total_tit') :
    blog_urls = i.attrs['href'] 
    blog_url.append(blog_urls)

print(f"블로그{len(blog_url)}개 의 주소입니다.")
print(blog_url)
```

```python
# 본문 크롤링
# Naver 는 본문이 iframe 안에 들어있음

contents = []

for i in blog_url:
    
    driver.get(i)
    time.sleep(1)
    # 블로그 안 본문에 있는 iframe에 접근
    driver.switch_to.frame("mainFrame")
    # 본문 내용 크롤링 시작
    try:
        a = driver.find_element_by_css_selector('div.se-main-container').text
        contents.append(a)
    # 만약 안되는 블로그가 있다면 예외 처리
    except NoSuchElementException:
        a = driver.find_element_by_css_selector('div#content-area').text
        contents.append(a)
        
# Selenium 창 닫기
driver.quit()
print('-------------')
print('|본문 크롤링 완료|')
print('-------------')
```

```python
df = pd.DataFrame({'제목':title,
                   '링크':blog_url,
                   '본문내용':contents,
    })


df
```
